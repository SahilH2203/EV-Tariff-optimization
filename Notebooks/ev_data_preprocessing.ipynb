## ACN Cleaning

```python
drop_cols = [
    "_meta",
    "end",
    "min_kWh",
    "start",
    "_items",
    "userInputs"
]
acn = acn.drop(columns=drop_cols)
```

## making ACN valuable

```python
acn_hourly = (
    acn
    .groupby(
        [
            "stationID",
            "hour"
        ]
    )
    .agg(
        demand_kwh=(
            "kWhDelivered",
            "sum"
        ),
        
        num_sessions=(
            "sessionID",
            "count"
        ),
        
        avg_session_kwh=(
            "kWhDelivered",
            "mean"
        ),
        
        avg_connected_hours=(
            "connected_hours",
            "mean"
        ),
        
        avg_charging_hours=(
            "charging_hours",
            "mean"
        )
    )
    .reset_index()
)
```

```python
print(acn_hourly.shape)
acn_hourly.head()
```

```python
#revenue
BASE_PRICE = 15
acn_hourly["price"] = BASE_PRICE
acn_hourly["revenue"] = (
    acn_hourly["demand_kwh"]
    * acn_hourly["price"]
)
```

```python
#station stats (capactiy)
station_stats = (
    acn
    .groupby("stationID")
    .agg(
        total_sessions=(
            "sessionID",
            "count"
        )
    )
)
```

## Temporal Feature Engineering

```python
acn_hourly["hour_of_day"] = (
    acn_hourly["hour"]
    .dt.hour
)
acn_hourly["day_of_week"] = (
    acn_hourly["hour"]
    .dt.dayofweek
)
acn_hourly["is_weekend"] = (
    acn_hourly["day_of_week"]
    >= 5
).astype(int)
acn_hourly["month"] = (
    acn_hourly["hour"]
    .dt.month
)
```

```python

print(
    acn_hourly["hour"].min()
)

print(
    acn_hourly["hour"].max()
)
acn_hourly.head()
acn_hourly.columns
```

```python
keep_cols = [
    "sessionID",
    "stationID",
    "connectionTime",
    "disconnectTime",
    "doneChargingTime",
    "kWhDelivered",
    "siteID",
    "clusterID"
]
acn.columns
```

```python
all_hours = pd.date_range(
    start=acn_hourly["hour"].min(),
    end=acn_hourly["hour"].max(),
    freq="H"
)
len(all_hours)
```

```python
all_stations = (
    acn_hourly["stationID"]
    .unique()
)
len(all_stations)
```

```python
# full grid of hour*stations
full_grid = pd.MultiIndex.from_product(
    [
        all_stations,
        all_hours
    ],
    names=[
        "stationID",
        "hour"
    ]
).to_frame(index=False)
full_grid.shape
```

```python
acn_final = (
    full_grid
    .merge(
        acn_hourly,
        on=[
            "stationID",
            "hour"
        ],
        how="left"
    )
)
```

```python
acn_final["hour_of_day"] = (
    acn_final["hour"]
    .dt.hour
)

acn_final["day_of_week"] = (
    acn_final["hour"]
    .dt.dayofweek
)

acn_final["is_weekend"] = (
    acn_final["day_of_week"] >= 5
).astype(int)

acn_final["month"] = (
    acn_final["hour"]
    .dt.month
)
```

```python
acn_final["region"] = "Caltech"

acn_final["data_source"] = "ACN"
```

## created urban_hourly data

```python
urban_hourly = (
    urban_long
    .groupby(
        [
            "grid",
            "hour"
        ]
    )
    .agg(
        demand_kwh=(
            "volume_kwh",
            "sum"
        ),

        avg_occupancy=(
            "occupancy",
            "mean"
        ),

        avg_duration_hours=(
            "duration_hours",
            "mean"
        ),

        avg_price=(
            "price",
            "mean"
        ),

        charger_count=(
            "count",
            "first"
        ),

        fast_count=(
            "fast_count",
            "first"
        ),

        slow_count=(
            "slow_count",
            "first"
        ),

        area=(
            "area",
            "first"
        ),

        lon=(
            "lon",
            "first"
        ),

        la=(
            "la",
            "first"
        ),

        CBD=(
            "CBD",
            "first"
        ),

        dynamic_pricing=(
            "dynamic_pricing",
            "first"
        )
    )
    .reset_index()
)
```

```python
urban_hourly["revenue"] = (
    urban_hourly["demand_kwh"]
    *
    urban_hourly["avg_price"]
)
```

```python
urban_hourly["utilization_rate"] = (
    urban_hourly["avg_occupancy"]
    /
    urban_hourly["charger_count"]
)
```

```python
urban_hourly['utilization_rate'].describe()
```

```python
urban_hourly["utilization_rate"] = (
    urban_hourly["utilization_rate"]
    .clip(
        lower=0,
        upper=1
    )
)
```

```python
urban_hourly["hour_of_day"] = (
    urban_hourly["hour"]
    .dt.hour
)

urban_hourly["day_of_week"] = (
    urban_hourly["hour"]
    .dt.dayofweek
)

urban_hourly["is_weekend"] = (
    urban_hourly["day_of_week"]
    >= 5
).astype(int)

urban_hourly["month"] = (
    urban_hourly["hour"]
    .dt.month
)
```
